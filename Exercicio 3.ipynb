```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

x = np.random.rand(1000)
w = np.random.randn(1000)
y = 1.2 + 2.3*x + 10*w

fig, ax = plt.subplots()
ax.plot(x, y, '.', color='black')
ax.set(xlabel='x', ylabel='y', title='Exerc√≠cio 3')
ax.grid(True)
plt.show()
```
